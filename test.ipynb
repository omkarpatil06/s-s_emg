```python
import matplotlib.pyplot as plt
```

```python
# Data for cutoff frequencies and corresponding accuracies
cutoff_freq = [50, 100, 150, 200, 250, 300, 350, 400, 450]
accuracy = [0.70270, 0.63063, 0.72072, 0.68468, 0.72072, 0.74775, 0.68468, 0.65765, 0.67567]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(cutoff_freq, accuracy, marker='o', linestyle='-', color='b', label='Accuracy')

# Add labels and title
plt.xlabel('Cutoff Frequency (Hz)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy vs. Cutoff Frequency', fontsize=14)
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
